```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
##add your web link 
web = requests.get("https://results.eci.gov.in/PcResultGenJune2024/index.htm")
```

```python
soup =BeautifulSoup(web.content, "html.parser")
```

```python
#for find any sessions& how many tables are here
tables= soup.find_all('table')
```

```python
tables
```

```python
if tables:
    table=tables[0]
    column_data = table.find_all('tr')
    headers = [header.text.strip() for header in column_data[0].find_all('th')]
    df=pd.DataFrame(columns =headers)
    for row in column_data[1:]:
        row_data = row.find_all('td')
        individual_row_data = [data.text.strip() for data in row_data]
        if len(individual_row_data) == len(headers):
            length = len(df)
            df.loc[length] = individual_row_data
        else:
            print("row does not match to headers")
else: 
    print("no table foubd")

print(df)
        
```

```python
#convert in csv or any file 'result you can put name according you'
df.to_csv('result.csv', index =False)
```
